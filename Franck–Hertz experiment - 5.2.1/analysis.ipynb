```python
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
data_4 = pd.read_csv('data/4.csv')
data_6 = pd.read_csv('data/6.csv')
data_8 = pd.read_csv('data/8.csv')
```

```python
trace1 = go.Scatter(
    x = data_4['V'],
    y = data_4['I'],
    mode = 'markers',
#     error_y = dict(
#         type = 'data',
#         array = data_sorted['err_W'],
#         visible = True,
#         color = '#000000',
#         thickness = 2,
#         width = 2
#     ),
#     error_x = dict(
#         type = 'constant',
#         value = 1.,
#         visible = True,
#         color = '#000000'
#     ),
    marker = dict(
        size = 10,
        opacity = 1,
        color = '#000000'
    )
)

trace2 = go.Scatter(
    x = [19, 19],
    y = [0, 90],
    line = dict(
        color = ('#ff0000'),
        width = 2,
        dash = 'dash'),
    showlegend = False
)

trace3 = go.Scatter(
    x = [38, 38],
    y = [0, 90],
    line = dict(
        color = ('#ff0000'),
        width = 2,
        dash = 'dash'),
    showlegend = False
)

trace0 = [trace1, trace2, trace3]

layout = go.Layout(
    title = "$Graph~1$",
    width = 1200,
    height = 1200,
    font = dict(
        size = 30,
        color = '#000000'
        ),
    showlegend=False,
    paper_bgcolor = '#ffffff',
    xaxis = dict(
        title = '$V_a,~V$',
        zeroline = False,
        range = [0, 55],
        ticklen = 0,
        color = '#000000',
        mirror = True,
        showline = True,
        linewidth = 2,
        titlefont = dict(
            size = 40,
            color = '#000000' 
            )
    ),
    yaxis = dict(
        title = '$I_k,~μA$',
        zeroline = False,
        range = [0, 90],
        ticklen = 0,
        color = '#000000',
        mirror = True,
        showline = True,
        linewidth = 2,
        titlefont = dict(
            size = 40,
            color = '#000000' 
            )
    ),
    margin = dict(
        l = 120,
        r = 23,
        t = 70,
        b = 100,
        pad = 0 
    )
)

fig = go.Figure(data=trace0, layout=layout)
#py.iplot(figure, filename='basic-error-bar')
py.image.save_as(fig, 'my_plot1.png')
#! open my_plot.png
from IPython.display import Image
Image("my_plot1.png", width=500)
```
